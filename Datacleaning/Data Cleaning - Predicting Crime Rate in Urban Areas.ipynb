```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../Raw Data/raw Data for crime anylasis.csv')
df.head(3)
```

# Data Cleaning

```python
df1=df.copy()
df1.isnull().sum()
```

```python
df1.drop(['index'],axis=1,inplace=True)
df1
```

```python
df1.drop_duplicates(inplace=True)
```

```python
cols_with_percent = [
    col for col in df1.columns
    if df1[col].astype(str).str.contains('%', case=False, na=False).any()
]

cols_with_Km =[
    col for col in df.columns
    if df[col].astype(str).str.contains('Km | KM | Km', case=False, na=False).any()
]

cols_with_nan = df.columns[df.isnull().any()].tolist()


print('Columns with percent:', cols_with_percent)
print('Columns with Km:', cols_with_Km)
print('Columns with NAN:',cols_with_nan)
```

```python
for i in cols_with_percent:
    df1[i] = df1[i].str.replace('%', '')
    df1[i] = df1[i].astype(float)
   
for i in cols_with_Km:
    df1[i] = df1[i].str.replace('km', '') 
    df1[i] = df1[i].astype(float)
```

```python
df1.head(5)
```

```python
for i in df1.columns:
    df1[i]=round(df1[i].fillna(df1[i].mean()),1)
```
